```python
y= ad_data['Clicked on Ad']
x=ad_data.drop(['Clicked on Ad','Ad Topic Line','Country','Timestamp','City',],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression as Lr
from sklearn.model_selection import train_test_split as tts
```

```python
adDATA = Lr(solver='liblinear')
```

```python
xtrain, xtest, ytrain, ytest = tts(x,y,test_size=0.3,random_state=101)
```

```python
adDATA.fit(xtrain,ytrain)
```

```python
pridict = adDATA.predict(xtest)
```

```python
from sklearn.metrics import classification_report as cReport
from sklearn.metrics import confusion_matrix as cFReport
```

```python
print(cReport(ytest,pridict))
```

```python
cFReport(ytest,pridict)
```

```python
from sklearn import metrics
metrics.accuracy_score(pridict,ytest)
```
