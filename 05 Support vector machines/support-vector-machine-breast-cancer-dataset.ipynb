```python
cancer = lbc()
```

```python
cancer.keys()
```

```python
df_feats = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df_feats.info()
```

```python
cancer['target_names']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x=df_feats
y=cancer['target']
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.svm import SVC
mod=SVC()
```

```python
mod.fit(xtrain,ytrain)
```

```python
preds  = mod.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix as cMat, classification_report as cFReport
```

```python
print(cMat(ytest,preds),'\n')
print(cFReport(ytest,preds))
```

```python
from sklearn.model_selection import GridSearchCV as GSCV
```

```python
param_grid = {'C':[0,1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GSCV(SVC(),param_grid,verbose=3)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_preds = grid.predict(xtest)
```

```python
print(cMat(ytest,grid_preds),'\n')
print(cFReport(ytest,grid_preds))
```
